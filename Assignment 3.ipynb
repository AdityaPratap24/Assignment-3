```python
#1.1
def myreduce(func,values):
    result = values[0]
    for i in values[1:]:
        result = func(result,i)
    return result
```

```python
def func(a,b):
    return a + b
```

```python
myreduce(func,[1,2,3,4,5,6])

```

```python
#1.2
def myfilter1(func,values):
    c = []
    for i in values:
        if func(i):
             c.append(i)
    return c
```

```python
def func(a):
    if type(a) == str:
        return True
    else:
        return False
```
